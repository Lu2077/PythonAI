```python
import pandas as pd

# === Configura ===
archivo_template = "2601141525_C_robot_translate_template_v3.7.0.xlsx" # Added .xlsx extension
hoja = 0
COL_VALUE = "detail" # Changed from "value" to "detail" to match the actual column name
COL_VALUE_CUSTOM = "detail(customize)" # Changed from "value(customize)" to "detail(customize)" to match the actual column name

# Diccionario con tus traducciones (puedes pegar cientos de entradas)
trad = {
    "Floor scrubber": "Fregadora de suelos",
    "Dust push brush": "Cepillo empujador de polvo",
    "Mopping brush": "Cepillo de fregado",
    "Squeegee rubber strip": "Banda de goma de la escobilla",
    "Front rubber strip of floor cleaning brush": "Banda de goma frontal del cepillo de limpieza de suelos",
    "Coverage speed": "Velocidad de cobertura",
    "Edge speed": "Velocidad de borde",
    "Is this mission closed?": "¿Está cerrada esta misión?",
    "Please drain and add fresh water": "Por favor, drene el agua y agregue agua limpia",
    "Continuation of the mandate": "Continuación de la tarea",
    "Clear water and sewage treated": "Agua limpia y aguas residuales procesadas",
    "The water has been replenished.": "El agua ha sido rellenada.",
    "Go to Drainage Location": "Ir a la ubicación de drenaje",
    "Sewage discharged": "Aguas residuales descargadas",
    "Go to Hydration Location": "Ir a la ubicación de llenado de agua",
    "Please add water.": "Por favor, agregue agua.",
}

df = pd.read_excel(archivo_template, sheet_name=hoja, engine="openpyxl")

# Validaciones
for col in [COL_VALUE, COL_VALUE_CUSTOM]:
    if col not in df.columns:
        raise ValueError(f"Falta la columna '{col}'. Columnas: {df.columns.tolist()}")

# Normalizar 'detail' y aplicar traducciones exactas
def normaliza(s):
    return "" if pd.isna(s) else str(s).strip()

df[COL_VALUE_CUSTOM] = df.apply(
    lambda r: trad.get(normaliza(r[COL_VALUE]), r.get(COL_VALUE_CUSTOM)),
    axis=1
)

# Guardar
salida = "keenon_template_ES.xlsx"
df.to_excel(salida, index=False)
salida
```
