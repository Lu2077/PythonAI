```python
#TAREA 3 EVALUACION 2
import networkx as nx
from queue import PriorityQueue

# Definir las rutas entre las ciudades
routes = [
    ("Santiago", "Valparaíso", 120, 20000),
    ("Santiago", "Rancagua", 90, 15000),
    ("Valparaíso", "La Serena", 400, 50000),
    ("Rancagua", "Talca", 250, 30000),
    ("La Serena", "Coquimbo", 10, 2000),
    ("Coquimbo", "Antofagasta", 900, 100000),
    ("Talca", "Chillán", 120, 18000),
    ("Chillán", "Concepción", 100, 15000),
    ("Concepción", "Temuco", 200, 25000),
    ("Temuco", "Valdivia", 120, 20000),
    ("Valdivia", "Osorno", 90, 17000),
    ("Osorno", "Puerto Montt", 110, 22000),
    ("Temuco", "Puerto Montt", 320, 50000),
    ("Concepción", "Los Ángeles", 127, 18000),
    ("Los Ángeles", "Temuco", 100, 17000),
    ("Talca", "Los Ángeles", 227, 35000),
    ("Valparaíso", "Los Andes", 50, 10000),
    ("Los Andes", "Mendoza (Arg.)", 450, 60000),
    ("Santiago", "Los Andes", 80, 15000),
    ("Osorno", "Bariloche (Arg.)", 120, 40000)
]

# Crear el grafo
G = nx.Graph()
for route in routes:
    G.add_edge(route[0], route[1], distance=route[2], cost=route[3])

# Función heurística (distancia en kilómetros)
def heuristic(node, goal):
    try:
        return nx.shortest_path_length(G, source=node, target=goal, weight='distance')
    except nx.NetworkXNoPath:
        return float('inf')

# Algoritmo A* (proporcionado sin modificar)
def a_star_search(graph, start, goal):
    open_set = PriorityQueue()
    open_set.put((0, start))
    came_from = {}
    cost_so_far = {start: 0}

    while not open_set.empty():
        current_cost, current = open_set.get()

        if current == goal:
            break

        for neighbor in graph.neighbors(current):
            distance = graph[current][neighbor]['distance']
            cost = graph[current][neighbor]['cost']
            new_cost = cost_so_far[current] + distance + cost

            if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                cost_so_far[neighbor] = new_cost
                priority = new_cost + heuristic(neighbor, goal)
                open_set.put((priority, neighbor))
                came_from[neighbor] = current

    return reconstruct_path(came_from, start, goal)

# Reconstrucción del camino
def reconstruct_path(came_from, start, goal):
    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from.get(current, None)
        if current is None:
            return []  # No hay camino
    path.append(start)
    path.reverse()
    return path

# Ejecución del algoritmo
start_city = "Santiago"
goal_city = "Puerto Montt"
path = a_star_search(G, start_city, goal_city)

# Mostrar resultados
if path:
    print("Ruta encontrada:", " -> ".join(path))
else:
    print(f"No se encontró una ruta entre {start_city} y {goal_city}.")
```

```python
#TAREA 1 EVALUACION 2

def dfs(grafo, inicio, visitados):
    pila = [inicio]
    componentes = []
    while pila:
        nodo = pila.pop()
        if nodo not in visitados:
            visitados.add(nodo)
            componentes.append(nodo)
            pila.extend(vecino for vecino in grafo[nodo] if vecino not in visitados)
    return componentes

def encontrar_componentes(grafo):
    visitados = set()
    todos_componentes = []
    for nodo in grafo:
        if nodo not in visitados:
            componentes = dfs(grafo, nodo, visitados)
            todos_componentes.append(componentes)
    return todos_componentes

# Grafo
grafo = {
    0: [1],
    1: [0, 2],
    2: [1],
    3: [4],
    4: [3],
    5: [],
}


componentes = encontrar_componentes(grafo)
print("Componentes encontrados:", componentes)
```
