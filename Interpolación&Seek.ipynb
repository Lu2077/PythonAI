```python
#grafico con interpolación
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d


hrs = [1,2,3,4]
T = []

a = float(input('ingrese T: '))
T.append(a)

b = float(input('ingrese T: '))
T.append(b)

c = float(input('ingrese T: '))
T.append(c)

d = float(input('ingrese T: '))
T.append(d)

print(T)

plt.plot(hrs, T,'bo',
          label='datos')
plt.title('hrs vs temperatura')

f = interp1d(hrs, T, kind='cubic')
hrs_new = np.linspace(1,4,100)
T_new = f(hrs_new)

plt.plot(hrs_new,T_new,'r',label='interpolación')
plt.xlabel('hrs')
plt.ylabel('T')
plt.grid(True)
plt.show()
```

```python

def busqueda_lineal(LISTA,Q):
        for e in LISTA:
            if e == Q:
                return True
        return False

LISTA = [1,2,3,4,5,7,9,10]

while True:
    try:
        Q = int(input("ingresa un número entero: "))

        if busqueda_lineal(LISTA,Q):
            print(f"el número {Q} es parte de la lista")
        else:
            print(f"el número {Q} no es parte de la lista")
        break

    except ValueError:
            print(" No es un número entero \n por favor vuelve a intentarlo")
```
